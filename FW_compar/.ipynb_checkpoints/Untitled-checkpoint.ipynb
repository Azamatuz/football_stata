```python

# Font Imports

import matplotlib.font_manager as fm
heading_font = fm.FontProperties(fname='/home/azamat/dev/jupyter/football_stata/PlayfairDisplay-Regular.ttf', size=28)
subtitle_font = fm.FontProperties(fname='/home/azamat/dev/jupyter/football_stata/Roboto-Regular.ttf', size=18)
lable_font = fm.FontProperties(fname='/home/azamat/dev/jupyter/football_stata/Roboto-Regular.ttf', size=20)

fontdict={'fontsize': 24,
          'weight' : 'bold',
         'horizontalalignment': 'center'}

sns.set_context("poster", font_scale=0.8)
```

```python
import pandas as pd
df_phys = pd.read_clipboard()
```

```python
df_phys
```

```python
plt.figure(figsize=(15,5))

x = df_phys['weight']
y = df_phys['hight']
labels = df_phys['name']

ax = sns.scatterplot(x=x, y=y, size=df_phys['age'])
for x_pos, y_pos, label in zip(x, y, labels):
    ax.annotate(label,             # The label for this point
                xy=(x_pos, y_pos), # Position of the corresponding point
                xytext=(7, 0),     # Offset text by 7 points to the right
                textcoords='offset points', # tell it to use offset points
                ha='left',         # Horizontally aligned to the left
                va='center')       # Vertical alignment is centered

plt.title('Comparison of FW\'s physique (height/weight)', fontproperties=heading_font, color="black",      
          position=(.28, 1.02))

plt.ylabel('Weight', fontproperties=lable_font)
plt.xlabel('Height', fontproperties=lable_font)
sns.despine(left=True)
```
