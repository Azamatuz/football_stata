```python
# Import charting libraries
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc("font", size=14)
import seaborn as sns
sns.set(style="white") #white background style for seaborn plots
sns.set(style="whitegrid", color_codes=True)
from math import pi

# Import data managment libraries
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
## Imoprt function 
#from data_uploader_from_csv.ipynb import data_uploader
pd.set_option('display.max_columns', None)
%matplotlib inline
sns.set()
```

```python
df = pd.read_csv('vsBene.csv', header=1)
df
```

```python
for key, name in enumerate(df.Player):
    if '-' in name:
        splitting = name.split('-')
        df.Player[key] = splitting[1]
    else: 
        df.Player[key] = name
df
```

```python
frames = [df[8:9], df[3:4]] 
df = pd.concat(frames)
df
```

```python
df_sum = df[['Player','Sh', 'SCA', 'Touches','Att', 'PrgDist', 'Carries','PrgDist.1', 'Press', 'Tkl', 'Int', 'Blocks']]
df_sum
```

```python
df_sum.columns = ['Player', 'Shots', 'SCA', 'Touches', 'Pass', 'PassPrgDist','Carries', 'CarryPrgDist', 'Press', 'Tackled', 'Interceptions', 'Blocks']
df_sum = df_sum.set_index('Player')
```

```python
df_trans = df_sum.T # or df1.transpose()

df_trans
```

```python
df_sum_copy = df_sum.copy()
df_sum_copy = df_sum_copy.astype(float).round(2)
df_sum_copy
```

```python
df_sum_copy.index[1]
```

```python

for key, val in enumerate(df_sum_copy):
    df_sum_copy[f'{val}'] = df_sum_copy[f'{val}']/df_sum_copy[f'{val}'].max()

df_sum_copy = df_sum_copy.round(2)
df_sum_copy
```

```python
# number of variable
categories=list(df_sum_copy)
N = len(categories)
categories
```

```python

 
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Initialise the spider plot
ax = plt.subplot(111, polar=True)
 
# If you want the first axis to be on top:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
 
# Draw one axe per variable + add labels labels yet
plt.xticks(angles[:-1], categories)
 
 # Draw ylabels
ax.set_rlabel_position(0)  
plt.yticks([.25, .5, .75], [".35",".50",".75"], color="grey", size=7)
plt.ylim(0,1)
 
# ------- PART 2: Add plots
 
# Plot each individual = each line of the data
# I don't do a loop, because plotting more than 3 groups makes the chart unreadable
 
# Ind1
values=df_sum_copy.iloc[0].values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label=df_sum_copy.index[0])
ax.fill(angles, values, 'b', alpha=0.1)
 
# Ind2
values=df_sum_copy.iloc[1].values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label=df_sum_copy.index[1])
ax.fill(angles, values, 'r', alpha=0.1)
 
# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.savefig((str(df_sum_copy.index[0] + 'VS' + df_sum_copy.index[1]) + ".png"))
```
