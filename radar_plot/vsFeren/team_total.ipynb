```python
frames = [prev_df[-1:], current_df[-1:]] 
df = pd.concat(frames)
df
```

```python
df_sum = df[['Sh', 'SCA', 'Touches', 'PrgDist', 'PrgDist.1', 'Press', 'Tkl', 'Int', 'Blocks']]
df_sum
```

```python
df_trans = df_sum.T # or df1.transpose()

df_trans
```

```python
df_sum_copy = df_sum.copy()
df_sum_copy = df_sum_copy.astype(float).round(2)
df_sum_copy
```

```python

for key, val in enumerate(df_sum_copy):
    df_sum_copy[f'{val}'] = df_sum_copy[f'{val}']/df_sum_copy[f'{val}'].max()

df_sum_copy = df_sum_copy.round(2)
df_sum_copy
```

```python
# number of variable
categories=list(df_sum_copy)
N = len(categories)
categories
```

```python

 
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Initialise the spider plot
ax = plt.subplot(111, polar=True)
 
# If you want the first axis to be on top:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
 
# Draw one axe per variable + add labels labels yet
plt.xticks(angles[:-1], categories)
 
 # Draw ylabels
ax.set_rlabel_position(0)  
plt.yticks([.25, .5, .75], [".35",".50",".75"], color="grey", size=7)
plt.ylim(0,1)
 
# ------- PART 2: Add plots
 
# Plot each individual = each line of the data
# I don't do a loop, because plotting more than 3 groups makes the chart unreadable
 
# Ind1
values=df_sum_copy.iloc[0].values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="vs Cagliare")
ax.fill(angles, values, 'b', alpha=0.1)
 
# Ind2
values=df_sum_copy.iloc[1].values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="vs Fener")
ax.fill(angles, values, 'r', alpha=0.1)
 
# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
```
