```python
df = pd.read_csv('PlayerStandardStats.csv', header=1)
df
```

```python
for key, name in enumerate(df.Player):
    splitting = name.split('\\')
    if len(splitting) > 0:
        df.Player[key] = splitting[1]
    else:
        df.Player[key] = splitting[0]
```

```python
for key, name in enumerate(df.Player):
    if '-' in name:
        splitting = name.split('-')
        df.Player[key] = splitting[1]
    else: 
        df.Player[key] = name
```
