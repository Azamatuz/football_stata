```python
# Font Imports

import matplotlib.font_manager as fm
heading_font = fm.FontProperties(fname='/home/azamat/dev/jupyter/football_stata/PlayfairDisplay-Regular.ttf', size=28)
subtitle_font = fm.FontProperties(fname='/home/azamat/dev/jupyter/football_stata/Roboto-Regular.ttf', size=18)
lable_font = fm.FontProperties(fname='/home/azamat/dev/jupyter/football_stata/Roboto-Regular.ttf', size=20)

fontdict={'fontsize': 24,
          'weight' : 'bold',
         'horizontalalignment': 'center'}

sns.set_context("poster", font_scale=0.8)
```

```python
df_shoot = pd.read_csv('shoot.csv', header=1)
df_shoot = df_shoot.fillna(0)

df_shoot.tail()
```

```python
plt.figure(figsize=(20,15))

y = df_shoot['Sh']
x = df_shoot['Gls']
labels = df_shoot['Squad']

sns.scatterplot(x=x, y=y )
ax = sns.regplot(x=x, y=y )


plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light'
)
#Juve highlighted
#plt.scatter(df.xG.where(df['Team']=='Juventus').dropna().values[0], df.xGA.where(df['Team']=='Juventus').dropna().values[0], color='r')

#plt.axhline(y=y.median(), color='gray', linestyle='-')
#plt.axvline(x=x.median(), color='gray', linestyle='-')
#ax = sns.scatterplot(x="Avg_temp", y="F_rate", hue="Class", data=temp_fert_2013)

for x_pos, y_pos, label in zip(x, y, labels):
    ax.annotate(label,             # The label for this point
                xy=(x_pos, y_pos), # Position of the corresponding point
                xytext=(7, 0),     # Offset text by 7 points to the right
                textcoords='offset points', # tell it to use offset points
                ha='left',         # Horizontally aligned to the left
                va='center')       # Vertical alignment is centered

ax.set_title('Comparison of shoots and goals, Seria A(2019/2020)', fontproperties=heading_font, color="black",         
             position=(0.3, 1.02));

plt.text(x=100, y=370, s='Data source: fbref.com',fontproperties=subtitle_font, horizontalalignment='right',color='#524939')

plt.ylabel('Total shoots', fontproperties=lable_font)
plt.xlabel('Goals', fontproperties=lable_font)
sns.despine(left=True)
```
