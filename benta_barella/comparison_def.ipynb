```python
df_benta = pd.read_csv('benta_def.csv', header=1)
df_benta = df_benta.fillna(0)
#df_benta = df_benta[['Min', 'Gls', 'Ast', 'Sh', 'Touches', 'Press', 'Blocks', 'xG', 'xA', 'SCA', 'GCA', 'Att', 'PrgDist', 'Carries', 'PrgDist.1', 'Succ', 'Att.1']]
#df_benta = df_benta.rename(columns={'Att': 'Pass_att', 'PrgDist': 'Pass_dist', 'PrgDist.1': 'Carr_dist', 'Succ': 'Drib_succ', 'Att.1': 'Drib_att'})
#df_benta['Min'][df_benta['Min']>90]=90
df_benta['Player'] = 'Bentancur'
df_benta.tail()
```

```python
df_barella = pd.read_csv('barella_def.csv', header=1)
df_barella = df_barella.fillna(0)
#df_benta = df_benta[['Min', 'Gls', 'Ast', 'Sh', 'Touches', 'Press', 'Blocks', 'xG', 'xA', 'SCA', 'GCA', 'Att', 'PrgDist', 'Carries', 'PrgDist.1', 'Succ', 'Att.1']]
#df_benta = df_benta.rename(columns={'Att': 'Pass_att', 'PrgDist': 'Pass_dist', 'PrgDist.1': 'Carr_dist', 'Succ': 'Drib_succ', 'Att.1': 'Drib_att'})
#df_benta['Min'][df_benta['Min']>90]=90
df_barella['Player'] = 'Barella'
df_barella.tail()
```

```python
df_barella_press = df_barella[['Press', 'Succ', '%', 'Def 3rd.1', 'Mid 3rd.1', 'Att 3rd.1', 'Player']]
df_barella_press = df_barella_press.drop(28)

df_barella_def3 = pd.DataFrame({'Val':df_barella_press['Def 3rd.1'].values, 'Zone': 'Def 3rd' }) 
df_barella_mid3 = pd.DataFrame({'Val':df_barella_press['Mid 3rd.1'].values, 'Zone': 'Mid 3rd' }) 
df_barella_att3 = pd.DataFrame({'Val':df_barella_press['Att 3rd.1'].values, 'Zone': 'Att 3rd' }) 

frames = [df_barella_def3, df_barella_mid3, df_barella_att3]
df_barella_3rd = pd.concat(frames)
df_barella_3rd['Player'] = 'Barella'
```

```python
df_benta_press = df_benta[['Press', 'Succ', '%', 'Def 3rd.1', 'Mid 3rd.1', 'Att 3rd.1', 'Player']]
df_benta_press = df_benta_press.drop(29)
df_benta_def3 = pd.DataFrame({'Val':df_benta_press['Def 3rd.1'].values, 'Zone': 'Def 3rd' }) 
df_benta_mid3 = pd.DataFrame({'Val':df_benta_press['Mid 3rd.1'].values, 'Zone': 'Mid 3rd' }) 
df_benta_att3 = pd.DataFrame({'Val':df_benta_press['Att 3rd.1'].values, 'Zone': 'Att 3rd' }) 
frames = [df_benta_def3, df_benta_mid3, df_benta_att3]
df_benta_3rd = pd.concat(frames)
df_benta_3rd['Player'] = 'Bentancur'
df_benta_3rd.tail()
```
