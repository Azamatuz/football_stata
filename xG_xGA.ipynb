```python
url = 'https://fbref.com/en/comps/11/Serie-A-Stats#all_stats_gca_squads'
```

```python
dframe_list = pd.io.html.read_html(url)
```

```python
dframe = dframe_list[1]
dframe
```

```python
dframe.columns.values
```

```python
df = pd.DataFrame()
df['Team'] = dframe['Unnamed: 1_level_0', 'Squad']
df['xG'] = dframe['Home', 'xG']
df['xGA'] = dframe['Home', 'xGA']
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc("font", size=14)
import seaborn as sns
sns.set(style="white") #white background style for seaborn plots
sns.set(style="whitegrid", color_codes=True)

import matplotlib.font_manager as fm
# Font Imports
heading_font = fm.FontProperties(fname='PlayfairDisplay-Regular.ttf', size=28)

subtitle_font = fm.FontProperties(fname='Roboto-Regular.ttf', size=18)

lable_font = fm.FontProperties(fname='Roboto-Regular.ttf', size=20)

# Color Themes
color_bg = '#FEF1E5'
darker_highlight = '#FBEADC'
lighter_highlight = '#ffffff'

plt.figure(figsize=(15,10))

fontdict={'fontsize': 24,
          'weight' : 'bold',
         'horizontalalignment': 'center'}

sns.set_context("poster", font_scale=0.8)

x = df['xG']
y = df['xGA']
labels = df['Team']

ax = sns.scatterplot(x=x, y=y, linewidths=2)

#Juve highlighted
plt.scatter(df.xG.where(df['Team']=='Juventus').dropna().values[0], df.xGA.where(df['Team']=='Juventus').dropna().values[0], color='r')

plt.axhline(y=y.median(), color='gray', linestyle='-')
plt.axvline(x=x.median(), color='gray', linestyle='-')
#ax = sns.scatterplot(x="Avg_temp", y="F_rate", hue="Class", data=temp_fert_2013)

for x_pos, y_pos, label in zip(x, y, labels):
    ax.annotate(label,             # The label for this point
                xy=(x_pos, y_pos), # Position of the corresponding point
                xytext=(7, 0),     # Offset text by 7 points to the right
                textcoords='offset points', # tell it to use offset points
                ha='left',         # Horizontally aligned to the left
                va='center')       # Vertical alignment is centered

ax.set_title('Seria A clubs, HOME xG and xGA (2019/2020)', fontproperties=heading_font, color="black",
         position=(0.3, 1.05));

plt.text(x=33, y=6, s='Data souce: fbref.com',fontproperties=subtitle_font, horizontalalignment='right',color='#524939')

plt.xlabel('Expected Goals(xG)', fontproperties=lable_font)
plt.ylabel('Expected Goals Allowed(xGA)', fontproperties=lable_font)
sns.despine(left=True)
```
